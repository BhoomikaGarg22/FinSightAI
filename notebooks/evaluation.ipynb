# FinSight AI - Evaluation Notebook

This notebook evaluates the FinSight AI Retrieval-Augmented Generation (RAG) pipeline.

The evaluation measures:

- Retrieval Success
- Response Relevance
- Response Latency

Author: Team FinSight AI

```python
import os
import sys
import time
import pandas as pd

# Add project root to Python path
project_root = os.path.abspath("..")
sys.path.append(project_root)

from backend.api import chat_with_ai
from backend.rag.retriever import get_retriever

print("Imports Successful!")
```

```python
retriever = get_retriever()

print("Retriever Loaded Successfully")
```

```python
questions = [

"What is the company's total revenue?",

"What are the major business risks?",

"What is the net profit?",

"What are the future growth opportunities?",

"What ESG initiatives are discussed?",

"What are the company's liabilities?",

"What are the main sources of revenue?",

"What investments were made during the year?",

"What are the management's future plans?",

"What are the key financial highlights?"

]
```

```python
results = []

for question in questions:

    start = time.time()

    # Retrieve documents
    docs = retriever.invoke(question)

    retrieval_success = len(docs) > 0

    # Generate answer
    answer = chat_with_ai(question)

    end = time.time()

    latency = round(end - start, 2)

    # Simple relevance score
    if len(answer) > 100:
        relevance = "High"
    elif len(answer) > 40:
        relevance = "Medium"
    else:
        relevance = "Low"

    results.append({

        "Question": question,

        "Retrieved Docs": len(docs),

        "Retrieval Success": retrieval_success,

        "Latency (s)": latency,

        "Response Relevance": relevance

    })
```

```python
df = pd.DataFrame(results)

df
```

```python
print("Average Latency :", round(df["Latency (s)"].mean(),2),"seconds")

print("Retrieval Rate :", round(df["Retrieval Success"].mean()*100,2),"%")

print("Average Retrieved Docs :", df["Retrieved Docs"].mean())
```
